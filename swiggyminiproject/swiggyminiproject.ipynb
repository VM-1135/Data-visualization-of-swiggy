```python
from scipy import stats
z=np.abs(stats.zscore(df['Price']))
z.sort_values()
```

```python
threshold_z= 8
outlier_indices=df[z>threshold_z]
```

```python
ded=df
```

```python
outlier_indices.index
```

```python
df.drop(outlier_indices.index,inplace=True)
df
```

```python
expensive=df[['Restaurant','Address','Price']].sort_values(by='Price',ascending=False).head(10)
```

```python
cheap=df[['Restaurant','Address','Price']].sort_values(by='Price',ascending=True).head(10)
```

```python
import seaborn as sns
```

```python
a= sns.barplot(x ='Restaurant',y = 'Price',data = expensive)
a.set_xticklabels(labels = df['Restaurant'].to_list(),rotation=90)
plt.show()
```
